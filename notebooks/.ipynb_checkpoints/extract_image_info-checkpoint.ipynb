## Extract info from image_statistics

```python
def processImageId(img_id):
    """
    Given an image id it adds zeroes to match the image_id of MS COCO

    # Arguments
        img_id: input id

    # Returns
        img_id: Id with leading zeros as string     
    """
    
    img_id = str(img_id)
    id_length = 12
    diff = id_length-len(img_id)

    if diff:
        img_id = "0"*diff+img_id
    return img_id
```

```python
all_mscoco_info = pd.read_csv('../data/image_statistics.csv', dtype={'ID': np.str})
```

```python
all_mscoco_info['ID'] = all_mscoco_info['ID'].apply(lambda x: processImageId(x))
```

```python
all_mscoco_info.head()
```

```python
our_image_info = all_mscoco_info.loc[all_mscoco_info['ID'].isin(image_order_1['image_id'])]
our_image_info['supercategories'] = our_image_info['supercategories'].apply(lambda x: x[2:-2])
our_image_info['categories'] = our_image_info['categories'].apply(lambda x: x[2:-2])
```

```python
our_image_info.columns = ['', 'brightness', 'rms_contrast', 'supercategories', 'categories',
       'ratio']
```
