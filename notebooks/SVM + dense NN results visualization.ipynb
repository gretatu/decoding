```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import xlrd
```

```python
def DataLoader(path, sheet):
    out = pd.read_excel(path, sheet_name=sheet)
    return out
```

```python
doc = xlrd.open_workbook('../data/svm_training_LOSO_RBF_kernel_std_overlap.xlsx').sheet_by_index(0)
```

```python
def getVals(doc_name):
    C_vals = doc_name.col_values(2)
    C_vals.pop(0)

    train_scores = doc_name.col_values(3)
    train_scores.pop(0)

    test_scores = doc_name.col_values(4)
    test_scores.pop(0)
    
    return C_vals, train_scores, test_scores
```
